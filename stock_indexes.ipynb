```python
indexes = ['ATX', 'BEL 20', 'BIRS', 'BSE SOFIX', 'CROBEX', 'Cyprus Main Market', 'PX', 'OMXC20', 'OMX Helsinki 25',
           'CAC 40', 'DAX', 'FTSE/Athex 20', 'Budapest SE', 'ICEX Main', 'ISEQ 20 Price', 'Italy 40', 'MSE',
          'MNSE 10', 'AEX', 'OSE Benchmark', 'WIG20', 'PSI All Share GR', 'BET', 'Belex 15', 'SAX',
           'Blue-Chip SBITOP', 'IBEX 35', 'OMXS30', 'SMI', 'BIST 100', 'PFTS', 'FTSE 100']
countries = ['Austria', 'Belgium', 'Bosnia', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark', 'Finland',
             'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 'Malta',
            'Montenegro', 'Netherlands', 'Norway', 'Poland', 'Portugal', 'Romania', 'Serbia', 'Slovakia',
            'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Turkey', 'Ukraine', 'United Kingdom']
appended_df = pd.DataFrame() 
n=0
x=0
for index in indexes:
    countries_names = []
    df = investpy.get_index_historical_data(index=indexes[n],
                                        country=countries[n],
                                        from_date='01/01/2010',
                                        to_date='01/12/2021',
                                        interval = 'Monthly')
    for country in range(0, len(df)):
        countries_names.append(countries[x])
    df['Country'] = countries_names
    appended_df = appended_df.append(df)
    n += 1
    x += 1
```
